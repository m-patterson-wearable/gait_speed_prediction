# Preparing unrounded speed  raw dataset (with outliers) for mlp

```python

#Importing raw (not cleaned) datset. The imported datset represent sets of continuous time series (ie walks) without any rounding of speed.
dataset=pd.read_csv("../data/processed/dataset_raw.csv",index_col=False )
dataset.head()
```

```python
# Construct MLP input dataset for with shape of no. of examles x (100x,100y,100z,speed, stamp of initial instance of acclearations , walk)
i=0
examples=[]
while i <= (len(dataset)-100):
    stamp=dataset['stamp'].iloc[i]
    walk=dataset['walk'].iloc[i]
    speed=dataset['speed'].iloc[i]
    accls=dataset[['x','y','z']].iloc[i:i+100].values.T.ravel()
    example=np.append(accls, [speed,stamp,walk])
    examples.append(example) #each example consist of 100x, 100y, 100x, 100rms, speed, stamp,walk
    i+=100
mlp_dataset=pd.DataFrame(examples)
mlp_dataset.to_csv("../data/processed/mlp_dataset_raw_xyz.csv", index=False)
```

```python
# Construct MLP input dataset for with shape of no. of examles x (100rms,speed, stamp of initial instance of acclearations , walk)
i=0
examples=[]
while i <= (len(dataset)-100):
    stamp=dataset['stamp'].iloc[i]
    walk=dataset['walk'].iloc[i]
    speed=dataset['speed'].iloc[i]
    rmss=dataset['rms'].iloc[i:i+100].values
    example=np.append(rmss, [speed,stamp,walk])
    examples.append(example) #each example consist of 100x, 100y, 100x, 100rms, speed, stamp,walk
    i+=100
mlp_dataset=pd.DataFrame(examples)
mlp_dataset.to_csv("../data/processed/mlp_dataset_raw_rms.csv", index=False)
```
